```python
#definition of the cubic deviation metric for scalars
import math
def CubicDeviationMetric(a,b,c,alpha,beta,gamma):
    if alpha<90:
        al=180-alpha
    else:
        al=alpha
    if beta<90:
        be=180-beta
    else:
        be=beta
    if gamma<90:
        ga=180-gamma
    else:
        ga=gamma
    alIN=round(math.cos(math.radians(al)),6) #rounding is necessary to avoid near zero small numbers instead of 0 when angle=90
    beIN=round(math.cos(math.radians(be)),6)
    gaIN=round(math.cos(math.radians(ga)),6)
    CDM=abs((a/math.sqrt(a**2+b**2-2*a*b*gaIN))-(1/math.sqrt(2)))+abs((b/math.sqrt(a**2+b**2-2*a*b*gaIN))-(1/math.sqrt(2)))+abs((a/math.sqrt(a**2+c**2-2*a*c*beIN))-(1/math.sqrt(2)))+abs((c/math.sqrt(a**2+c**2-2*a*c*beIN))-(1/math.sqrt(2)))+abs((c/math.sqrt(b**2+c**2-2*b*c*alIN))-(1/math.sqrt(2)))+abs((b/math.sqrt(b**2+c**2-2*b*c*alIN))-(1/math.sqrt(2)))
    return(CDM)
```

```python
#example of use for scalars
CubicDeviationMetric(1,2,3,60,120,72)
```

```python
#load in example data and calculate the metric for these data
#although it's possible to do this in many different ways, we use a lambda function so the same metric function works on scalars and datasets
import pandas
data = pandas.read_csv("example data.csv")
data['CDM']=data.apply(lambda x: CubicDeviationMetric(x['a'], x['b'], x['c'], x['alpha'], x['beta'], x['gamma']), axis=1)
data
```
